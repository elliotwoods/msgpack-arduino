```python
from cobs import cobs
from math import ceil, floor

plain_text_length = 300

plain_text = [1] * plain_text_length
plain_binary = bytearray(plain_text)
encoded=cobs.encode(plain_binary)
encoded_length = len(encoded)

print(plain_text_length)
print(encoded_length)

estimated_length = floor(plain_text_length/254) + 1 + plain_text_length
print(estimated_length)
```
